```python
import numpy as np

"""
Menghasilkan fitur GLCM.

Masukan:
F = Citra berskala keabuan

Keluaran:
Fitur = fitur GLCM untuk beberapa sudut.
"""

tinggi, lebar = len(grayImage), len(grayImage[0])

# Bentuk GLCM 
GLCM0 = np.zeros((256, 256))
total_piksel0 = 0

GLCM45 = np.zeros((256, 256))
total_piksel45 = 0

GLCM90 = np.zeros((256, 256))
total_piksel90 = 0

GLCM135 = np.zeros((256, 256))
total_piksel135 = 0

for y in range(1, tinggi-1): 
    for x in range(1, lebar-1):
        # Sudut 0
        a = grayImage[y][x]
        b = grayImage[y][x+1]
        GLCM0[a+1][b+1] += 1
        total_piksel0 += 1

        # Sudut 45
        a = grayImage[y][x]
        b = grayImage[y-1][x+1]
        GLCM45[a+1][b+1] += 1
        total_piksel45 += 1

        # Sudut 90
        a = grayImage[y][x]
        b = grayImage[y-1][x]
        GLCM90[a+1][b+1] += 1
        total_piksel90 += 1

        # Sudut 135
        a = grayImage[y][x]
        b = grayImage[y-1][x-1]
        GLCM135[a+1][b+1] += 1
        total_piksel135 += 1

GLCM0 = GLCM0 / total_piksel0
GLCM45 = GLCM45 / total_piksel45
GLCM90 = GLCM90 / total_piksel90
GLCM135 = GLCM135 / total_piksel135

# Hitung ASM
asm0 = 0.0
asm45 = 0.0
asm90 = 0.0
asm135 = 0.0
for a in range(255):
    for b in range(255):
        asm0 = asm0 + (GLCM0[a+1, b+1] * GLCM0[a+1, b+1])
        asm45 = asm45 + (GLCM45[a+1, b+1] * GLCM45[a+1, b+1])
        asm90 = asm90 + (GLCM90[a+1, b+1] * GLCM90[a+1, b+1])
        asm135 = asm135 + (GLCM135[a+1, b+1] * GLCM135[a+1, b+1])


# Hitung kontras
kontras0 = 0
kontras45 = 0
kontras90 = 0
kontras135 = 0

for a in range(255):
    for b in range(255):
        kontras0 += ((a - b)**2) * GLCM0[a+1, b+1]
        kontras45 += ((a - b)**2) * GLCM45[a+1, b+1]
        kontras90 += ((a - b)**2) * GLCM90[a+1, b+1]
        kontras135 += ((a - b)**2) * GLCM135[a+1, b+1]

# Hitung IDM
idm0 = 0
idm45 = 0
idm90 = 0
idm135 = 0

for a in range(255):
    for b in range(255):
        idm0 += GLCM0[a+1, b+1] / (1 + (a - b)**2)
        idm45 += GLCM45[a+1, b+1] / (1 + (a - b)**2)
        idm90 += GLCM90[a+1, b+1] / (1 + (a - b)**2)
        idm135 += GLCM135[a+1, b+1] / (1 + (a - b)**2)

# hitung entropi
  
entropi0 = 0.0
entropi45 = 0.0
entropi90 = 0.0
entropi135 = 0.0

for a in range(255):
    for b in range(255):
        if GLCM0[a+1][b+1] != 0:
            entropi0 -= GLCM0[a+1][b+1] * math.log(GLCM0[a+1][b+1])
            
        if GLCM45[a+1][b+1] != 0:
            entropi45 -= GLCM45[a+1][b+1] * math.log(GLCM45[a+1][b+1])
            
        if GLCM90[a+1][b+1] != 0:
            entropi90 -= GLCM90[a+1][b+1] * math.log(GLCM90[a+1][b+1])
            
        if GLCM135[a+1][b+1] != 0:
            entropi135 -= GLCM135[a+1][b+1] * math.log(GLCM135[a+1][b+1])
          
# Hitung kovarians
# Hitung px [] dan py [] dulu
korelasi0 = 0.0
px0 = 0
py0 = 0
reratax0 = 0.0
reratay0 = 0.0
stdevx0 = 0.0
stdevy0 = 0.0
  
korelasi45 = 0.0
px45 = 0
py45 = 0
reratax45 = 0.0
reratay45 = 0.0
stdevx45 = 0.0
stdevy45 = 0.0

korelasi90 = 0.0
px90 = 0
py90 = 0
reratax90 = 0.0
reratay90 = 0.0
stdevx90 = 0.0
stdevy90 = 0.0
    
korelasi135 = 0.0
px135 = 0
py135 = 0
reratax135 = 0.0
reratay135 = 0.0
stdevx135 = 0.0
stdevy135 = 0.0
        
for a in range(255):
    for b in range(255):
        px0 = px0 + a * GLCM0[a+1, b+1]  
        py0 = py0 + b * GLCM0[a+1, b+1]
            
        px45 = px45 + a * GLCM45[a+1, b+1] 
        py45 = py45 + b * GLCM45[a+1, b+1]
            
        px90 = px90 + a * GLCM90[a+1, b+1] 
        py90 = py90 + b * GLCM90[a+1, b+1]
            
        px135 = px135 + a * GLCM135[a+1, b+1]  
        py135 = py135 + b * GLCM135[a+1, b+1]

#Hitung deviasi standar
for a in range(255):
  for b in range(255):
      stdevx0 = stdevx0 + (a-px0) * (a-px0) * GLCM0[a+1, b+1]
      stdevy0 = stdevy0 + (b-py0) * (b-py0) * GLCM0[a+1, b+1]
          
      stdevx45 = stdevx45 + (a-px45) * (a-px45) * GLCM45[a+1, b+1]                            
      stdevy45 = stdevy45 + (b-py45) * (b-py45) * GLCM45[a+1, b+1]
                            
          
      stdevx90 = stdevx90 + (a-px90) * (a-px90) * GLCM90[a+1, b+1]                              
      stdevy90 = stdevy90 + (b-py90) * (b-py90) * GLCM90[a+1, b+1]
                            
          
      stdevx135 = stdevx135 + (a-px135) * (a-px135) * GLCM135[a+1, b+1]                                
      stdevy135 = stdevy135 + (b-py135) * (b-py135) * GLCM135[a+1, b+1]

# Hitung korelasi
for a in range(255):
    for b in range(255):
        korelasi0 = korelasi0 + ((a-px0)*(b-py0)* GLCM0[a+1, b+1]/(stdevx0*stdevy0))
            
        korelasi45 = korelasi45 + ((a-px45)*(b-py45)* GLCM45[a+1, b+1]/(stdevx45 * stdevy45))
            
        korelasi90 = korelasi90 + ((a-px90)*(b-py90)* GLCM90[a+1, b+1]/(stdevx90 * stdevy90))
            
        korelasi135 = korelasi135 + ((a-px135)*(b-py135)* GLCM135[a+1, b+1]/(stdevx135 * stdevy135))
            
```

```python
# Calculate the co-occurrence matrix for the image
co_matrix = skimage.feature.graycomatrix(grayImage, [5], [0, np.pi/2, np.pi/4], levels=256, symmetric=False, normed=True)

# Calculate texture features from the co-occurrence matrix
contrast = skimage.feature.graycoprops(co_matrix, 'contrast')
correlation = skimage.feature.graycoprops(co_matrix, 'correlation')
energy = skimage.feature.graycoprops(co_matrix, 'energy')
homogeneity = skimage.feature.graycoprops(co_matrix, 'homogeneity')
dissimilarity = skimage.feature.graycoprops(co_matrix, 'dissimilarity')
#entropy = skimage.feature.graycoprops(co_matrix, 'entropy')

# Print the texture features
print("Contrast:", contrast)
print("Correlation:", correlation)
print("Energy:", energy)
print("Homogeneity:", homogeneity)
#print("Entropy :", entropy)
print("dissimilarity :", dissimilarity)
```
